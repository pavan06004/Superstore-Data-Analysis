## Import libraries

```python
import pandas as pd
import numpy as np
```

## Load the cleaned dataset

```python
df = pd.read_csv("~/Downloads/Superstore_dataset.csv")
```

## Check Dataset

```python
df.head()
```

## Convert date columns

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

## Create Year

```python
df["Year"] = df["Order Date"].dt.year
```

```python
df[["Order Date","Year"]].head()
```

## Create Month

```python
df["Month"] = df["Order Date"].dt.month_name()
```

```python
df[["Order Date", "Month"]].head()
```

## Create Quarter

```python
df["Quarter"] = df["Order Date"].dt.quarter
```

```python
df[["Order Date", "Quarter"]].head()
```

## Create Shipping Days

```python
df["Shipping Days"] = (df["Ship Date"] - df["Order Date"]).dt.days
```

```python
df[["Order Date", "Ship Date", "Shipping Days"]].head()
```

## Create Profit Margin

```python
df["Profit Margin"] = (df["Profit"] / df["Sales"]) * 100
```

```python
df[["Sales", "Profit", "Profit Margin"]].head()
```

## Create Average Selling Price

```python
df["Average Selling Price"] = df["Sales"] / df["Quantity"]
```

```python
df[["Sales", "Quantity", "Average Selling Price"]].head()
```

## Create Discount Category

```python
df["Discount Category"] = pd.cut(
    df["Discount"],
    bins=[-0.01, 0, 0.2, 0.5, 1],
    labels=["No Discount", "Low", "Medium", "High"]
)
```

```python
df[["Discount", "Discount Category"]].head()
```

## Create Sales Category

```python
df["Sales Category"] = pd.cut(
    df["Sales"],
    bins=[0, 100, 500, 1000, 25000],
    labels=["Low", "Medium", "High", "Very High"]
)
```

```python
df[["Sales", "Sales Category"]].head()
```

## Create Order Size

```python
df["Order Size"] = pd.cut(
    df["Quantity"],
    bins=[0, 2, 5, 14],
    labels=["Small", "Medium", "Large"]
)
```

```python
df[["Quantity", "Order Size"]].head()
```

## Verify All New Columns

```python
df.head()
```

## Save the Feature-Engineered Dataset

```python
df.to_csv("feature_engineered_superstore.csv", index=False)
```
