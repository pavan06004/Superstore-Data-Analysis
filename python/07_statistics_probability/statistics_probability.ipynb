# Statistics and Probability Analysis

## Import Libraries


```python
import pandas as pd
import numpy as np
```

## Load Feature-Engineered Dataset

```python
df = pd.read_csv("../06_feature_engineering/feature_engineered_superstore.csv")
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

## Descriptive Statistics

```python
df.describe()
```

## Mean

```python
print("Mean Sales:", df["Sales"].mean())
print("Mean Profit:", df["Profit"].mean())
print("Mean Quantity:", df["Quantity"].mean())
```

## Median

```python
print("Median Sales:", df["Sales"].median())
print("Median Profit:", df["Profit"].median())
```

## Mode

```python
print(df["Ship Mode"].mode())
print(df["Category"].mode())
```

## Variance

```python
print("Sales Variance:", df["Sales"].var())
print("Profit Variance:", df["Profit"].var())
```

## Standard Deviation

```python
print("Sales Std Dev:", df["Sales"].std())
print("Profit Std Dev:", df["Profit"].std())
```

## Quartiles

```python
print(df["Sales"].quantile([0.25, 0.50, 0.75]))
```

## Interquartile Range (IQR)

```python
Q1 = df["Sales"].quantile(0.25)
Q3 = df["Sales"].quantile(0.75)

IQR = Q3 - Q1

print("IQR:", IQR)
```

## Skewness

```python
print("Sales Skewness:", df["Sales"].skew())
print("Profit Skewness:", df["Profit"].skew())
```

## Correlation

```python
df[["Sales", "Profit", "Quantity", "Discount"]].corr()
```

# Probability Examples

## Probability that an order belongs to the Technology category

```python
technology_probability = (
    df["Category"] == "Technology"
).mean()

print("P(Technology):", technology_probability)
```

## Probability that an order is from the West region

```python
west_probability = (
    df["Region"] == "West"
).mean()

print("P(West):", west_probability)
```

## Conditional Probability

Probability that an order is Technology, given that it is from the West region:

```python
west_orders = df[df["Region"] == "West"]

conditional_probability = (
    west_orders["Category"] == "Technology"
).mean()

print("P(Technology | West):", conditional_probability)
```
